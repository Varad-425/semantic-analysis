```python
import pandas as pd
import nltk
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset=pd.read_csv('twitter_training.csv')
test=pd.read_csv('twitter_validation.csv')
```

```python
print(dataset)
```

```python
xds=dataset[dataset['Borderlands']=='Borderlands']
```

```python
X_ds=xds.iloc[:,3].values
Y_ds=xds.iloc[:,2].values
```

```python
print(Y_ds)
print(X_ds)
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, len(X_ds)):
  review = re.sub('[^a-zA-Z]', ' ', str(X_ds[i]))
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
print(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray() # type: ignore
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
ysy=lr.fit_transform(Y_ds)
```

```python
print(lr.inverse_transform([0,1,2,3]))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,ysy,test_size=0.20,random_state=0)
print(x_train)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
y_pred=gb.predict(x_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
acs=accuracy_score(y_test,y_pred)
print(acs)
```
